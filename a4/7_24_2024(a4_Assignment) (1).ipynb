```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
white_wine = pd.read_csv('winequality-white.csv', sep = ';')
white_wine.head()
```

```python
#1. For each feature and the target variables, obtain the mean, standard deviation, min, max, and 25/50/75% percentiles. (Show a table.)
white_wine.describe()
```

```python
#2. Illustrate the correlation matrix of all features and the target.
sns.heatmap(white_wine.corr(),annot=True, annot_kws= {'size':5} ,cmap='coolwarm')
plt.title('White Wine Correlation Matrix')
plt.show()
```

```python
#3. Plot a histogram of the target value (quality).
white_wine['quality'].unique()
white_wine.hist(column='quality', bins=7, color='orange') #added 7 bins to clarify histogram
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.title('Distribution of Wine Quality')
```

```python
#These are all on different scales
#Need to standardize the scale between all features so that it is centered around zero and variance is scaled to standard deviation
wine_data_standardized = (white_wine - white_wine.mean()) / white_wine.std()
# Create violin plots for each feature
sns.violinplot(data=wine_data_standardized, inner='quartile', scale='width')
plt.xticks(rotation=90) #rotate so visible
plt.title('White Wine Features Distribution (Standardized)')
plt.figure(figsize=(15, 10))
```

```python
#5. Train a linear regression model with a single input variable (alcohol). Plot all data points (x=alcohol; y=quality) and the obtained regression model. Show the intercept, coefficient, RMSE, and R^2.
#linear regression model with alcohol

plt.scatter(x=white_wine['alcohol'], y=white_wine['quality'], color='black')
model = LinearRegression()
model.fit(white_wine[['alcohol']], white_wine['quality'])

#Requirements: intercept, coeeficent, RMSE, R2, Regression line
intercept = model.intercept_
coefficient = model.coef_[0]
fitted = model.predict(white_wine[['alcohol']])
rmse = np.sqrt(mean_squared_error(white_wine['quality'], fitted))
r2 = r2_score(white_wine['quality'], fitted)
print(f"Intercept = {intercept}")
print(f"Coefficient = {coefficient}")
print(f"RMSE = {rmse}")
print(f"R^2 = {r2}")

# Regression line
plt.plot(white_wine['alcohol'], fitted, color='red',label='Regression line')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.title('Linear Regression Model Alcohol vs. Quality')
```

```python
#6. Train a linear regression model with input variables (volatile_acidity, alcohol, and density). Show the intercept, coefficient, RMSE, and R^2.
# Training model with volatile acidity, alcohol, and density
model = LinearRegression()
model.fit(white_wine[['volatile acidity', 'alcohol', 'density']], white_wine['quality'])
#REquirements: same as above problem, but for all variable
intercept = model.intercept_
coefficients = model.coef_
fitted = model.predict(white_wine[['volatile acidity', 'alcohol', 'density']])
rmse = np.sqrt(mean_squared_error(white_wine['quality'], fitted))
r2 = r2_score(white_wine['quality'], fitted)
print(f'Intercept = {intercept}')
print(f'Coefficients = {coefficients}')
print(f'RMSE = {rmse}')
print(f'R^2 = {r2}')
#pairplots to visualize the relationship between the multiple features/variables
relevant_columns = ['volatile acidity', 'alcohol', 'density', 'quality']
sns.pairplot(white_wine[relevant_columns], kind='reg', diag_kind='kde', plot_kws={'scatter_kws': {'color': 'black'}, 'line_kws': {'color': 'red'}})
```
